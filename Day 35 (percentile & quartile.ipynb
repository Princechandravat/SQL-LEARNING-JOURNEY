# PERCENTILE,QUARTILE,SKEW

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Prince Chandravat\OneDrive\Videos\Documents\DATASETS\stock market dataset\BRITANNIA.csv").dropna()
data.head()
```

```python
#mean-
mean=data['Deliverable Volume'].mean()
print('mean:',mean)

#median-
median=data['Deliverable Volume'].median()
print('median:',median)

#standard deviation-
std=data['Deliverable Volume'].std()
print('std:',std)

#Variance
variance=data['Deliverable Volume'].var()
print('variance:',variance)
```

```python
#PERCENTILE -

Q1=np.percentile(data['Deliverable Volume'],25)  # 25th percentile (q1)
Q2=np.percentile(data['Deliverable Volume'],50)  # 50th percentile (q2)
Q3=np.percentile(data['Deliverable Volume'],75)  # 75th percentile (q3)

#IQR-
iqr=Q3-Q1  #iqr (q3-q1)

#min/max-
minimum=Q1-1.5*iqr
maximum=Q3+1.5*iqr

outliers=[x for x in data['Deliverable Volume'] if x<minimum or x>maximum]
```

```python
print('Q1:',Q1)  #q1
print('Q2:',Q2)  #q2
print('Q3:',Q3)  #q3
print('IQ:R',iqr)  #iqr
print('minimum:',minimum)  #minimum
print('maximum:',maximum)   #maximum
print('outliers:',outliers)  #outliers
```

```python
#boxplot-

plt.figure(figsize=(10,5))
plt.boxplot(data['Deliverable Volume'],vert=False)
plt.title("Boxplot for Outlier Detection")
plt.xlabel("Deliverable Volume")
plt.show()
```
